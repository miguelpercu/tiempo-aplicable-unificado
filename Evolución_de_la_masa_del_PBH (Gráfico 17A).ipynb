```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

def generar_datos_masa(t_total_sim=8.5e6, num_puntos=100):
    tiempo_simulacion = np.linspace(0, t_total_sim, num_puntos)
    masa_inicial = 1e8  # kg
    masa_final = 9.999e7  # kg
    masa = np.linspace(masa_inicial, masa_final, num_puntos)
    return tiempo_simulacion, masa

def plot_evolucion_masa(tiempo, masa):
    plt.figure(figsize=(8, 6))
    plt.plot(tiempo, masa, label="Masa del PBH", color="blue")
    plt.xlabel("Tiempo de Simulaci√≥n ($t_{event}$) (s)")
    plt.ylabel("Masa del PBH ($M$) (kg)")
    plt.title("Evoluci√≥n de la Masa del PBH (Gr√°fico 17A)")
    ax = plt.gca()
    ax.xaxis.set_major_formatter(ticker.ScalarFormatter(useMathText=True))
    ax.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
    ax.yaxis.set_major_formatter(ticker.ScalarFormatter(useMathText=True))
    ax.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.tight_layout()
    plt.savefig("grafico_17A_evolucion_masa_pbh.png")
    print("Gr√°fico 'grafico_17A_evolucion_masa_pbh.png' guardado.")

if __name__ == "__main__":
    tiempo_sim, masa_pbh = generar_datos_masa(t_total_sim=8.5e6)
    plot_evolucion_masa(tiempo_sim, masa_pbh)
```
